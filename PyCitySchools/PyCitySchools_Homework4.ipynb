```python
import pandas as pd
```

```python
students_csv = "Resources/students_complete.csv"
schools_csv = "Resources/schools_complete.csv"

students_data = pd.read_csv(students_csv)
schools_data = pd.read_csv(schools_csv)
```

```python
students_data.head()
```

```python
schools_data.head()
```

```python
district_data = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
district_data.head()
```

```python
district_df = pd.DataFrame(district_data)
district_df.head()
```

 ## District Summary

```python
total_schools = schools_data["School ID"].count()
total_schools

total_students = students_data["student_name"].count()
total_students

total_budget = schools_data["budget"].sum()
total_budget

avg_math = district_df["math_score"].mean()
avg_math

avg_read = district_df["reading_score"].mean()
avg_read

percent_math = district_df.loc[district_df["math_score"] >= 70]
percent_math

good_math = (len(percent_math)/total_students) * 100
good_math

percent_read = district_df.loc[district_df["reading_score"] >= 70]
percent_read

good_read = (len(percent_read)/total_students) * 100
good_read

good_overall = (good_math + good_read)/2
good_overall
```

```python
District_Summary = [{"School Total": total_schools,
                                "Student Total": total_students,
                                "Total Budget": total_budget,
                                "Average Math Score": avg_math,
                                "Average Reading Score": avg_read,
                                "% Passing Math": good_math,
                                "% Passing Reading": good_read,
                                "Overall Passing %": good_overall}]
District_Summary_df = pd.DataFrame(District_Summary)
District_Summary_df
```

## School Summary

```python
schools_df = pd.DataFrame(district_data)
schools_df = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
schools_df
```

```python
schools_sort_df = schools_df.sort_values("school_name", ascending = True)
schools_sort_df
```

```python
schools_index = schools_data.set_index("school_name")
schools_index
```

```python
school_type = schools_index["type"]
school_type
```

```python
school_total_students = schools_index["size"]
school_total_students
```

```python
school_budget = schools_index["budget"]
school_budget
```

```python
Budg_per_Stud = school_budget / school_total_students
Budg_per_Stud
```

```python
schools_group_df = schools_sort_df.groupby(["school_name"])
schools_group_df.head()
```

```python
school_avg_math = schools_group_df["math_score"].mean()
school_avg_math
```

```python
school_avg_read = schools_group_df["reading_score"].mean()
school_avg_read
```

```python
passing_math = schools_sort_df[schools_sort_df["math_score"] >= 70]
passing_math
```

```python
totalmath_pass_df = passing_math.groupby(["school_name"]).count()["student_name"]
totalmath_pass_df
```

```python
passing_reading = schools_sort_df[schools_sort_df["reading_score"] >= 70]
passing_reading.head()
```

```python
totalreading_pass_df = passing_reading.groupby(["school_name"]).count()["student_name"]
totalreading_pass_df
```

```python
pctpass_math = totalmath_pass_df / school_total_students * 100
pctpass_math
```

```python
pctpass_reading = totalreading_pass_df / school_total_students * 100
pctpass_reading
```

```python
overall_pass_df = schools_sort_df[(schools_sort_df["math_score"] >= 70) & (schools_sort_df["reading_score"] >= 70)]
overall_pass_df
```

```python
overall_passing = overall_pass_df.groupby(["school_name"]).count()["student_name"]
overall_passing
```

```python
pctoverall_pass = (overall_passing / school_total_students) * 100
pctoverall_pass
```

```python
Schools_Summary = ({"School Type": school_type,
                    "Total Students": school_total_students,
                    "Total School Budget": school_budget,
                    "Budget per Student": Budg_per_Stud,
                    "Average Math Score": school_avg_math,
                    "Average Reading Score": school_avg_read,
                    "% Passed Math": pctpass_math,
                    "% Passed Reading": pctpass_reading,
                    "% Overall Passed": pctoverall_pass,
})

Schools_Summary_df = pd.DataFrame(Schools_Summary)
Schools_Summary_df
```

# Scores by School Spending

```python
bins = [0, 585, 630, 645, 680]
labels = ["<585", "585-630", "630-645", "645-680"]
```

# Scores by School Size

```python
bins = [0, 1000, 2000, 5000]
size_labels = ["Small(<1000)", "Medium(1000-2000)", "Large(2000-5000)"]

Schools_Summary_df["Student Size"] = pd.cut(Schools_Summary_df["Total Students"], bins, labels = size_labels)
Size_Summary = Schools_Summary_df.groupby("Student Size")
Size_Summary = Size_Summary[["Average Math Score", "Average Reading Score", "% Passed Math", "% Passed Reading", "% Overall Passed"]]
Size_Summary.mean()
```
